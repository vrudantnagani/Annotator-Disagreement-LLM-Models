### Original Dataset

```python
import pandas as pd
import collections
import operator
```

```python
file_path = 'ucberkeley_measuring_hate_speech_dataset_preprocessed.csv'
df = pd.read_csv(file_path)
```

```python
anno_ids = df.annotator_id
len(anno_ids)
```

```python
an_id_count = collections.Counter(anno_ids)
# printing the element and the frequency
for key, value in an_id_count.items():
    print(f"{key}: {value}")
```

```python
comm_ids = df.comment_id.unique()
len(comm_ids)
```

```python
comm_ids = df.comment_id
com_id_count = collections.Counter(comm_ids)
# printing the element and the frequency
for key, value in com_id_count.items():
    print(f"{key}: {value}")
```

```python
sorted_d = dict(sorted(com_id_count.items(), key=operator.itemgetter(1),reverse=True))
sorted_d
```

```python
res = sum(x == 1 for x in com_id_count.values())
res
```

```python
lst = []
for key, value in sorted_d.items():
    if value == 1:
        lst.append(key)
lst
```

```python
df = df[df.comment_id.isin(lst) == False]
```

```python
comm_ids = df.comment_id.unique()
len(comm_ids)
```

```python
df.to_csv('ucberkeley_measuring_hate_speech_dataset_preprocessed_v2.csv', index = False)
```
